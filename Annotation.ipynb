```python
targetFiles = 'smallData'
# targetFiles = 'HealthyFood'
# targetFiles = 'UnhealthyFood'
# targetFiles = 'GenericTweets'
RAW_CSV_DIR = f'./data/CSVFiles/'
PROCESSED_CSV_FILE = f'./{RAW_CSV_DIR}{targetFiles}-AnnotatedDataframe.csv'
```

```python
import pandas as pd
import numpy as np
import os
import math
```

```python
chunksize = 10000
# tfr = pd.read_csv('./TempFiles/'+ healthy + 'DataFrame.csv', usecols=['id', 'text', 'full_text', 'user'], index_col=None, header=0, lineterminator='\n',chunksize=chunksize, iterator=True)
# df_healthy = pd.concat(tfr, ignore_index=True)
file = f'{RAW_CSV_DIR}{targetFiles}-DataFrame.csv'
tfr = pd.read_csv(file, encoding='utf-8', usecols=['id', 'text', 'full_text', 'user', 'truncated', 'text2Proc'], index_col=None, header=0, lineterminator='\n',chunksize=chunksize, iterator=True)
df_annotation = pd.concat(tfr, ignore_index=True)
df_annotation.head()
```

```python
#Get Username

users = df_annotation.user.values
usernames = []

for user in users:
    keyword = "'screen_name':"
    before_keyword, keyword, after_keyword = user.partition(keyword)
    my_string = after_keyword
    keyword = "'location':"
    before_keyword, keyword, after_keyword = my_string.partition(keyword)
    username = before_keyword.replace(',','').replace("\'",'').replace('\"','').strip()
    usernames.append(username)
```

```python
df_annotation = df_annotation.assign(username = usernames)
df_annotation = df_annotation.replace('None', np.NaN)
df_annotation.info()
```

```python
df_annotation = df_annotation.sample(n=500)
df_annotation = df_annotation[['id','text2Proc', 'username']]
df_annotation
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
target = []

for index, row in df_annotation.iterrows():
    print(color.BOLD + '@{} - Tweet: {} '.format(row['username'], row['text2Proc']) + color.END)
    val = int(input("Talkinf about ...? No related=0, Consumption=1, Desire=2, Recommendation=3, Ads=4, Other=5: "))
    print("")
    target.append(val)
    
    
```

```python
df_annotation['target'] = target
df_annotation
```

```python
df_annotation.to_csv(PROCESSED_CSV_FILE, index= False)
```
