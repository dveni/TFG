```python
#Libraries used in the notebook
import csv
import json
import numpy as np
import pandas as pd
import glob
import os
import ntpath
import pickle
import urllib.request as url
```

```python
targetFiles = 'smallData'
# targetFiles = 'HealthyFood'
# targetFiles = 'UnhealthyFood'
# targetFiles = 'GenericTweets'
RAW_CSV_DIR = f'./data/CSVFiles/'
TARGET_CSV_FILE = f'{RAW_CSV_DIR}{targetFiles}-DataFrame.csv'
```

```python
chunksize = 10000
tfr = pd.read_csv(TARGET_CSV_FILE , usecols = ['id','id_str','lang','place','text', 'extended_entities'],
                  index_col=None, header=0, lineterminator='\n',chunksize=chunksize, iterator=True)
df = pd.concat(tfr, ignore_index=True)
```

```python
i = df.extended_entities[1]
a = i.replace("'", '"')
b = json.loads(a)
b

```

```python
def load_extended_entities(x):
    
    
    try:
        y = json.loads(x.replace("'", '"'))
    except:
        y = None
    return y
```

```python
df['extended_entities_parsed'] = df.extended_entities.apply(load_extended_entities)
df
```

```python
def get_pics_urls(x):
    
    urls = {}
    
    if x:
        print('Media legnth: {}'.format(len(x['media'])))
        for element in x['media']:
            if element['type']=='photo':
                urls[element['id_str']]=(element['media_url'])
    return urls
```

```python
df['pics_urls'] = df.extended_entities_parsed.apply(get_pics_urls)
df
```

```python
def download_pics(x):
    
    
    if x['pics_urls']:
        outdir = './data/images/' + str(x['id_str'])
        if not os.path.exists(outdir):
            os.mkdir(outdir)
            for element in x['pics_urls']:
            
                outname = '{}'.format(element + '.jpg')
                fullname = os.path.join(outdir, outname)
                print('Downloading {}: {} to {}'.format(element, x['pics_urls'][element], fullname))
                try:
                    url.urlretrieve(x['pics_urls'][element], fullname)
                except:
                    print('Error during downloading {}: {} to {}'.format(element, x['pics_urls'][element], fullname))
        
        
            
    
```

```python
df.apply(download_pics, axis=1)
```
