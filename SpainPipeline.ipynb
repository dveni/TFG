```python
# import geopandas as gpd
# import geoplot
import pandas as pd
import json
import numpy as np

targetFiles = 'UnhealthyFood'
```

```python
chunksize = 10000
tfr = pd.read_csv('./data/CSVFiles/SpainUnhealthyFoodDataframe.csv',
                  index_col=None, header=0, lineterminator='\n',chunksize=chunksize, iterator=True)
df = pd.concat(tfr, ignore_index=True)
```

```python
#Get full text from extended tweet
def getFullText(truncated, extended_tweet):
    if truncated:
        a = extended_tweet.replace("'", '"')
        try:
            b = json.loads(a)
            res = b['full_text']
        except:
            res = None
        return res
```

```python
df['full_text'] = df.apply(lambda x: getFullText(x['truncated'], x['extended_tweet']) , axis=1)
```

```python
def getText2Proc(fullText, text):
    if fullText is not None:
        return (fullText)
    else:
        return (text)
```

```python
# Sacar text2proc
df['text2Proc'] = df.apply(lambda x: getText2Proc(x['full_text'], x['text']) , axis=1)
```

```python
#Get @usernames
users = df.user.values
usernames = []
for user in users:
    keyword = "'screen_name':"
    before_keyword, keyword, after_keyword = user.partition(keyword)
    my_string = after_keyword
    keyword = "'location':"
    before_keyword, keyword, after_keyword = my_string.partition(keyword)
    username = before_keyword.replace(',','').replace("\'",'').replace('\"','').strip()
    usernames.append(username)
```

```python
#Get names
names = []
for user in users:
    keyword = "'name':"
    before_keyword, keyword, after_keyword = user.partition(keyword)
    my_string = after_keyword
    keyword = "'screen_name':"
    before_keyword, keyword, after_keyword = my_string.partition(keyword)
    username = before_keyword.replace(',','').replace("\'",'').replace('\"','').strip()
    names.append(username)
```

```python
endpoint = 'http://senpy.gsi.upm.es/api'
import requests
from IPython.display import Code

def query(endpoint, **kwargs):
    '''Query a given Senpy endpoint with specific parameters, and prettify the output'''
    res = requests.get(endpoint,
                       params=kwargs)
    if res.status_code != 200:
        raise Exception(res)
    return Code(res.text, language=kwargs.get('outformat', 'json-ld'))
```
